```python
import numpy as np
import pandas as pd
import datetime
import scipy.stats as stats
from matplotlib import pyplot as plt
%matplotlib inline
```

```python
# datetime.strptime
datelist = pd.date_range('1-1-2016', '1-2-2016', freq='1H')
data = np.random.randint(100,150,size=np.size(datelist))

elapsed_time = []

for i in datelist:
    if np.size(elapsed_time) > 0:
        deltaT = i - datelist[0]
        deltaT = deltaT.total_seconds()
        elapsed_time = np.append(elapsed_time, deltaT)
    else:
        elapsed_time = 0
        

slope, intercept, rVal, pVal, stderr = stats.linregress(elapsed_time, data)

fit = slope*elapsed_time + intercept
np.size(fit)
```

```python
fig = plt.figure(figsize=(6,4))
ax = plt.gca()
plt.subplots_adjust(bottom=0.17,left=0.17,top=0.96,right=0.96)

# Change the axis units to serif
plt.setp(ax.get_ymajorticklabels(),fontsize=12)
plt.setp(ax.get_xmajorticklabels(),fontsize=12)

ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')

ax.xaxis.set_ticks_position('bottom')
ax.yaxis.set_ticks_position('left')

# Turn on the plot grid and set appropriate linestyle and color
ax.grid(True,linestyle=':',color='0.75')
ax.set_axisbelow(True)

# Define the X and Y axis labels
plt.xlabel('Time', fontsize=22,weight='bold',labelpad=5)
plt.ylabel('Data', fontsize=22,weight='bold',labelpad=10)

plt.plot(datelist, data, linewidth=2,linestyle='--',label=r'Data')
plt.plot(datelist, fit, linewidth=2,linestyle='-',label=r'Regression')



# uncomment below and set limits if needed
# plt.xlim(0,30)
plt.ylim(75, 175)

# Create the legend, then fix the fontsize
leg = plt.legend(loc='best', ncol = 2, fancybox=True)
ltext  = leg.get_texts()
plt.setp(ltext,fontsize=18)

# Adjust the page layout filling the page using the new tight_layout command
plt.tight_layout(pad=0.5)

# plt.savefig('Delay Responses/response_tau_{}.pdf'.format(tau_string))

plt.show()
plt.close()
```
